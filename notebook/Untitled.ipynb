```python
fig, ax = plt.subplots(1,2,figsize=(16,6))
levels=np.linspace(0,4)
cm="pink_r"
dir = "/home/smoon/Dropbox/gc/out"
prob = "GC_test"
t0 = 0
t1 = 90
hN=32
ds = read(dir,prob,t0)
ct=ax[0].contourf(ds['x'],ds['y'],np.log10(ds['den'][hN,:,:]),levels=levels,cmap=cm)
ax[0].streamplot(ds['x'],ds['y'],ds['vel'][hN,:,:,0],ds['vel'][hN,:,:,1],color='y')
fig.colorbar(ct,ax=ax[0])
ds = read(dir,prob,t1)
ct=ax[1].contourf(ds['x'],ds['y'],np.log10(ds['den'][hN,:,:]),levels=levels,cmap=cm)
ax[1].streamplot(ds['x'],ds['y'],ds['vel'][hN,:,:,0],ds['vel'][hN,:,:,1],color='y')
fig.colorbar(ct,ax=ax[1])
ax[0].set_xlim(-500,500)
ax[0].set_ylim(-500,500)
ax[1].set_xlim(-500,500)
ax[1].set_ylim(-500,500)
```
