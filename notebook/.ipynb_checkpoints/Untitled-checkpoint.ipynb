```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import astropy.constants as c
import astropy.units as u
import sys
sys.path.insert(0,'../Athena-TIGRESS/python/')
import pyathena as pa
def read(dir, i):
    """ read vtk and return x, y, z, density, velocity """
    da = pa.AthenaDataSet(dir+"/id0/GC."+"{:04g}".format(i)+".vtk")
    xmin=da.domain['left_edge']
    xmax=da.domain['right_edge']
    dx=da.domain['dx']
    x=np.arange(xmin[0],xmax[0],dx[0])+0.5*dx[0]
    y=np.arange(xmin[1],xmax[1],dx[1])+0.5*dx[1]
    z=np.arange(xmin[2],xmax[2],dx[2])+0.5*dx[2]
    den = da.read_all_data('density')
    vel = da.read_all_data('velocity')
    T = da.read_all_data("temperature")
    cs = da.read_all_data("sound_speed")
    dic = {"x":x, "y":y, "z":z, "den":den, "vel":vel, "T":T, "cs":cs}
    return dic
def vcirc(x,y,z,omg=0):
    M0 = 1e10*u.Msun
    R0 = 300*u.pc
    R2 = (x[None,:]**2+y[:,None]**2)*u.pc**2
    v2 = c.G*M0*R2/(R2+R0*R0)**(3./2.)
    omg *= u.km/u.s/u.pc
    v = np.sqrt(v2)
    return (v - np.sqrt(R2)*omg).to(u.km/u.s)
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,6))
levels=np.linspace(1,2)
dir = "test"
t0 = 0
t1 = 50
ds = read(dir,t0)
ct=ax[0].contourf(ds['x'],ds['y'],np.log10(ds['den'][8,:,:]),levels=levels,extend="both")
ax[0].streamplot(ds['x'],ds['y'],ds['vel'][8,:,:,0],ds['vel'][8,:,:,1],color='y')
fig.colorbar(ct,ax=ax[0])
ds = read(dir,t1)
ct=ax[1].contourf(ds['x'],ds['y'],np.log10(ds['den'][8,:,:]),levels=levels,extend="both")
ax[1].streamplot(ds['x'],ds['y'],ds['vel'][8,:,:,0],ds['vel'][8,:,:,1],color='y')
fig.colorbar(ct,ax=ax[1])
ax[0].set_xlim(-500,500)
ax[0].set_ylim(-500,500)
ax[1].set_xlim(-500,500)
ax[1].set_ylim(-500,500)
```
