```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import astropy.constants as c
import astropy.units as u
import sys
from scipy.optimize import bisect

def KI02(T):
    return (1e7*np.exp(-114800/(T+1000)) + 0.14*np.sqrt(T)*np.exp(-92/T))*2e-26*u.erg/u.s*u.cm**3
def pe(T,U=1):
    if (T > 30):
        return 2e-26*U*u.erg/u.s
    else:
        return 2e-30*U*u.erg/u.s
def cr(zeta16):
    """eq. (30.3) in Draine"""
    return 1.03e-27*zeta16*u.erg/u.s
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,6))

U=[1e0,1e1,1e2]
for j in [0,1,2]:
    zeta16=0
    den = np.logspace(-5,5)
    Teq = np.zeros(len(den))
    for i,n in enumerate(den):
        n /= u.cm**3
        try:
            res = bisect(lambda T: (n*n*KI02(T)-n*(pe(T,U=U[j])+cr(zeta16))).value, 1e1,1e7)
            Teq[i] = res
        except:
            Teq[i] = np.nan
    Teq *= u.K
    den /= u.cm**3
    prs = den*Teq
    ax[0].loglog(den,Teq,'-',label="U="+str(U[j]))
    ax[1].loglog(den,prs,'-',label="U="+str(U[j]))
    
    zeta16=10
    den = np.logspace(-2,5)
    Teq = np.zeros(len(den))    
    for i,n in enumerate(den):
        n /= u.cm**3
        try:
            res = bisect(lambda T: (n*n*KI02(T)-n*(pe(T,U=U[j])+cr(zeta16))).value, 1e1,1e7)
            Teq[i] = res
        except:
            Teq[i] = np.nan
    Teq *= u.K
    den /= u.cm**3
    prs = den*Teq
    ax[0].loglog(den,Teq,'--')
    ax[1].loglog(den,prs,'--')
    
    
ax[0].legend()
ax[0].set_xlabel(r"$n_{\rm H}$")
ax[0].set_ylabel(r"$T_{\rm eq}$")
ax[1].set_xlabel(r"$n_{\rm H}$")
ax[1].set_ylabel(r"$P/k_{\rm B}$")
```
