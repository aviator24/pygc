# Rotating Frame Test
* Check initial equilibrium is well maintained in both rotating and non-rotating frame
* External potential is given by 
$$ \Phi = -\frac{GM}{\sqrt{R^2+R_0^2}}, $$
$$ g_x = -\frac{GMx}{(R^2+R_0^2)^{3/2}} $$
$$ g_y = -\frac{GMy}{(R^2+R_0^2)^{3/2}} $$
 with curresponding circular velocity
$$ v_{\rm circ} = \sqrt{\frac{GMR^2}{(R^2+R_0^2)^{3/2}}}. $$
* $M = 10^{10}\,M_\odot$ and $R_0 = 300\,{\rm pc}$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import astropy.constants as c
import astropy.units as u
import sys
import os
sys.path.insert(0,'../Athena-TIGRESS/python/')
from matplotlib.colors import LogNorm
import pyathena as pa
def ds(dir, i):
    """ read vtk and return x, y, z, density, velocity """
    da = pa.AthenaDataSet(dir+"/id0/GC."+"{:04g}".format(i)+".vtk")
    xmin=da.domain['left_edge']
    xmax=da.domain['right_edge']
    dx=da.domain['dx']
    x=np.arange(xmin[0],xmax[0],dx[0])+0.5*dx[0]
    y=np.arange(xmin[1],xmax[1],dx[1])+0.5*dx[1]
    z=np.arange(xmin[2],xmax[2],dx[2])+0.5*dx[2]
    den = da.read_all_data('density').mean(axis=0)
    vel = da.read_all_data('velocity').mean(axis=0)
    return x,y,z,den,vel
def vcirc(x,y,z,omg=0):
    M0 = 1e10*u.Msun
    R0 = 300*u.pc
    R2 = (x[None,:]**2+y[:,None]**2)*u.pc**2
    v2 = c.G*M0*R2/(R2+R0*R0)**(3./2.)
    omg *= u.km/u.s/u.pc
    v = np.sqrt(v2)
    return (v - np.sqrt(R2)*omg).to(u.km/u.s)
def KimElm17(x,y,omg=0):
    R = np.sqrt((x[None,:]**2+y[:,None]**2))
    v = 215 + 95*np.tanh((R-70)/60) - 50*np.log10(R) + 1.5*(np.log10(R))**3
    return (v-R*omg)*u.km/u.s
```

## $c_s = 1\,{\rm km/s}$
## $\Omega = [0$, $0.2$, $1]$

```python
fig, ax = plt.subplots(3,3,figsize=(24,18))
levels=0.98 + np.arange(0,11)*0.004
dir0 = "../out/rotframe_gas_test/C1W0"
omg0 = 0
dir1 = "../out/rotframe_gas_test/C1W1"
omg1 = 0.2
dir2 = "../out/rotframe_gas_test/C1W2"
omg2 = 1
t0 = 147
ycut = 64
cmap="seismic"

x,y,z,den,vel = ds(dir0,t0)
ct=ax[0,0].contourf(x,y,den,levels=levels,cmap=cmap,extend="both")
ax[0,0].streamplot(x,y,vel[:,:,0],vel[:,:,1])
fig.colorbar(ct,ax=ax[0,0])
vc = vcirc(x,y,z,omg=omg0)
R = np.sqrt(x[None,:]**2 + y[:,None]**2)
ax[1,0].plot(x, (vc*x/R)[ycut,:])
ax[1,0].plot(x, vel[ycut,:,1])
ax[2,0].plot(x, den[ycut,:])

x,y,z,den,vel = ds(dir1,t0)
ct=ax[0,1].contourf(x,y,den,levels=levels,cmap=cmap,extend="both")
#ct.set_under("k")
#ct.set_over("w")
ax[0,1].streamplot(x,y,vel[:,:,0],vel[:,:,1])
fig.colorbar(ct,ax=ax[0,1])
vc = vcirc(x,y,z,omg=omg1)
R = np.sqrt(x[None,:]**2 + y[:,None]**2)
ax[1,1].plot(x, (vc*x/R)[ycut,:])
ax[1,1].plot(x, vel[ycut,:,1])
ax[2,1].plot(x, den[ycut,:])

x,y,z,den,vel = ds(dir2,t0)
ct=ax[0,2].contourf(x,y,den,levels=levels,cmap=cmap,extend="both")
ax[0,2].streamplot(x,y,vel[:,:,0],vel[:,:,1])
fig.colorbar(ct,ax=ax[0,2])
vc = vcirc(x,y,z,omg=omg2)
R = np.sqrt(x[None,:]**2 + y[:,None]**2)
ax[1,2].plot(x, (vc*x/R)[ycut,:])
ax[1,2].plot(x, vel[ycut,:,1])
ax[2,2].plot(x, den[ycut,:])

ax[0,0].set_xlim(-500,500)
ax[0,1].set_xlim(-500,500)
ax[0,2].set_xlim(-500,500)

ax[0,0].set_ylim(-500,500)
ax[0,1].set_ylim(-500,500)
ax[0,2].set_ylim(-500,500)

ax[1,0].set_xlim(-500,500)
ax[1,1].set_xlim(-500,500)
ax[1,2].set_xlim(-500,500)

ax[1,0].set_ylim(-250,250)
ax[1,1].set_ylim(-250,250)
ax[1,2].set_ylim(-250,250)

ax[2,0].set_xlim(-500,500)
ax[2,1].set_xlim(-500,500)
ax[2,2].set_xlim(-500,500)

ax[2,0].set_ylim(0.5,1.5)
ax[2,1].set_ylim(0.5,1.5)
ax[2,2].set_ylim(0.5,1.5)

ax[2,0].set_xlabel(r"$x$")
ax[1,0].set_ylabel(r"$v_y$")
ax[2,0].set_ylabel(r"$\rho$")
ax[0,0].set_title("lab frame")
ax[0,1].set_title("rotating frame, "+r"$\Omega=0.2\,{\rm km/s/pc}$")
ax[0,2].set_title("rotating frame, "+r"$\Omega=1\,{\rm km/s/pc}$")
```

# RotFrame test in Kim & Elmegreen (2017) rotation curve
## This is computed using direct acceleration source term, which will be deprecated, because potential is needed for star particle integrator.

## $$\Omega = [0, 0.04]$$

```python
fig, ax = plt.subplots(3,2,figsize=(16,18))
levels=0.98 + np.arange(0,11)*0.004
dir0 = "../out/rotframe_gas_test/C1W0"
omg0 = 0
dir1 = "../out/rotframe_gas_test/C1W1"
omg1 = 0.04
t0 = 10
ycut = 64
cmap="seismic"

x,y,z,den,vel = ds(dir0,t0)
ct=ax[0,0].contourf(x,y,den,levels=levels,cmap=cmap,extend="both")
ax[0,0].streamplot(x,y,vel[:,:,0],vel[:,:,1])
fig.colorbar(ct,ax=ax[0,0])
vc = KimElm17(x,y,omg=omg0)
R = np.sqrt(x[None,:]**2 + y[:,None]**2)
ax[1,0].plot(x, (vc*x/R)[ycut,:])
ax[1,0].plot(x, vel[ycut,:,1])
ax[1,0].plot(x, vel[ycut,:,1]/abs(x)*200)
ax[2,0].plot(x, den[ycut,:])

x,y,z,den,vel = ds(dir1,t0)
ct=ax[0,1].contourf(x,y,den,levels=levels,cmap=cmap,extend="both")
#ct.set_under("k")
#ct.set_over("w")
ax[0,1].streamplot(x,y,vel[:,:,0],vel[:,:,1])
fig.colorbar(ct,ax=ax[0,1])
vc = KimElm17(x,y,omg=omg1)
R = np.sqrt(x[None,:]**2 + y[:,None]**2)
ax[1,1].plot(x, (vc*x/R)[ycut,:])
ax[1,1].plot(x, vel[ycut,:,1])
ax[1,1].plot(x, vel[ycut,:,1]/abs(x)*200)
ax[2,1].plot(x, den[ycut,:])

ax[0,0].set_xlim(-500,500)
ax[0,1].set_xlim(-500,500)

ax[0,0].set_ylim(-500,500)
ax[0,1].set_ylim(-500,500)


ax[1,0].set_xlim(-500,500)
ax[1,1].set_xlim(-500,500)


ax[1,0].set_ylim(-250,250)
ax[1,1].set_ylim(-250,250)


ax[2,0].set_xlim(-500,500)
ax[2,1].set_xlim(-500,500)


ax[2,0].set_ylim(0.5,1.5)
ax[2,1].set_ylim(0.5,1.5)


ax[2,0].set_xlabel(r"$x$")
ax[1,0].set_ylabel(r"$v_y$")
ax[2,0].set_ylabel(r"$\rho$")
ax[0,0].set_title("lab frame")
ax[0,1].set_title("rotating frame, "+r"$\Omega=0.2\,{\rm km/s/pc}$")
```
